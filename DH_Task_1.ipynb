##**Problem Statement and Goal**

- **Problem Statement:** Perform an exploratory data analysis (EDA) on the Iris dataset to understand its characteristics.

- **Goal:** Load, inspect, summarize, and visualize the Iris dataset to identify data trends, distributions, and relationships between features.

- **IMPORT LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

- **DATA VISUALIZATION**

#### Scatter Plot: Relationships between Features

```python
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset Features', y=1.02) # Add a title to the pair plot
plt.show()
```

#### Histograms: Value Distributions

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle('Histograms of Iris Dataset Features', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
```

#### Box Plots: Identify Outliers

```python
plt.figure(figsize=(15, 10))

for i, feature in enumerate(iris.columns[:-1]): # Exclude the 'species' column
    plt.subplot(2, 2, i + 1) # Arrange plots in a 2x2 grid
    sns.boxplot(y=feature, x='species', data=iris, palette='viridis')
    plt.title(f'Box Plot of {feature} by Species')

plt.tight_layout()
plt.show()
```
